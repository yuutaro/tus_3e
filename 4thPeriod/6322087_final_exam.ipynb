```python
### **ÔºëÔºéÂãïÁîª„ÅÆË™≠„ÅøËæº„Åø**ÔºàÊúÄÂ§ß10ÁÇπÔºâ
# ÂãïÁîªÔºàPeople - 84973.mp4Ôºâ„ÇíË™≠„ÅøËæº„ÅøÔºå  ÂãïÁîª„ÅÆÁ∏¶„ÉªÊ®™„ÅÆ„Çµ„Ç§„Ç∫ÔºåFPSÔºåÁ∑è„Éï„É¨„Éº„É†Êï∞„ÇíÂá∫Âäõ„Åó„Å¶„Åè„Å†„Åï„ÅÑÔºé
import cv2

# „É≠„Éº„Ç´„É´Áí∞Â¢ÉÁî®
path = "People - 84973.mp4"

# ÊèêÂá∫(google colabÁí∞Â¢É)Áî®
# path = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/People - 84973.mp4"

# ÂãïÁîª„Éï„Ç°„Ç§„É´„ÇíË™≠„ÅøËæº„Åø
video = cv2.VideoCapture(path)

# ÂãïÁîª„ÅÆÁ∏¶„ÉªÊ®™„ÅÆ„Çµ„Ç§„Ç∫ÔºåFPSÔºåÁ∑è„Éï„É¨„Éº„É†Êï∞„ÇíÂèñÂæó
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = video.get(cv2.CAP_PROP_FPS)
frame_count = int(video.get(cv2.CAP_PROP_FRAME_COUNT))

# ÂèñÂæó„Åó„ÅüÊÉÖÂ†±„ÇíË°®Á§∫
print("ÂãïÁîª„ÅÆÂπÖ:", width)
print("ÂãïÁîª„ÅÆÈ´ò„Åï:", height)
print("FPS:", fps)
print("Á∑è„Éï„É¨„Éº„É†Êï∞:", frame_count)

# „É°„É¢„É™„ÇíËß£Êîæ
video.release()
```

ÂãïÁîª„ÅÆ„Éï„Ç°„Ç§„É´ÂêçÔºàPeople - 84973.mp4Ôºâ„ÅÆÊñáÂ≠óÂàó„ÇíÂá¶ÁêÜ„Åó„Å¶Ôºå„Çø„Ç§„Éà„É´ÔºàPeopleÔºâ„Å®IDÔºà84973Ôºâ„Å´ÂàÜ„Åë„Çã„Éó„É≠„Ç∞„É©„É†„Çí‰ΩúÊàê„ÅóÔºå„Åù„ÅÆÁµêÊûú„ÇíÂá∫Âäõ„Åó„Å¶„Åè„Å†„Åï„ÅÑÔºé

```python
### **ÔºìÔºé„Éï„Ç°„Ç§„É´Âêç„ÅÆ„Éë„Éº„Çπ**ÔºàÊúÄÂ§ß10ÁÇπÔºâ

import os

# „É≠„Éº„Ç´„É´Áí∞Â¢ÉÁî®
path1 = "People - 84973.mp4"
path2 = "Cat - 66004.mp4"
path3 = "Flowers - 66823.mp4"
path4 = "Dubrovnik - 12866.mp4"

# ÊèêÂá∫(google colabÁí∞Â¢É)Áî®
# path = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/People - 84973.mp4"
# path2 = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/Cat - 66004.mp4"
# path3 = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/Flowers - 66823.mp4"
# path4 = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/Dubrovnik - 12866.mp4"

# „Éï„Ç°„Ç§„É´Âêç„ÇíÂàÜÂâ≤„Åô„ÇãÈñ¢Êï∞
def parse_filename(path):
    file_name = os.path.basename(path)
    title, id = file_name.split(" - ")
    print("Title:", title)
    id = id.split(".")[0]
    print("ID:", id)
    print()

# Âá∫Âäõ
parse_filename(path1)
parse_filename(path2)
parse_filename(path3)

```

### ÔºîÔºé‰∫åÂÄ§Âåñ„Å®ÂãïÁîª„ÅÆ‰øùÂ≠òÔºàÊúÄÂ§ß10ÁÇπÔºâ

ÂãïÁîªÔºàCat - 66004.mp4Ôºâ„ÇíÔºíÂÄ§Âåñ„Åó„ÅüÂãïÁîª„Çí‰ΩúÊàê„Åô„Çã„Éó„É≠„Ç∞„É©„É†„Çí‰ΩúÊàê„ÅóÔºå„Åù„ÅÆÂãïÁîª„Çí‰øùÂ≠ò„Åó„Å¶„Åè„Å†„Åï„ÅÑÔºéÈñæÂÄ§„ÅØËá™ÂàÜ„ÅßË®≠ÂÆö„Åó„Å¶„Åã„Åæ„ÅÑ„Åæ„Åõ„ÇìÔºé

‰∫åÂÄ§ÂåñÔºöÁîªÁ¥†ÂÄ§„ÅåÈñæÂÄ§„Çà„ÇäÂ§ß„Åç„Åë„Çå„Å∞ÁôΩ„ÇíÂâ≤„ÇäÂΩì„Å¶Ôºå„Åù„ÅÜ„Åß„Å™„Åë„Çå„Å∞Èªí„ÇíÂâ≤„ÇäÂΩì„Å¶„ÇãÂá¶ÁêÜ

<aside>
‚ö†Ô∏è Á∏¶„ÉªÊ®™„ÅÆ„Çµ„Ç§„Ç∫„ÅØÂÖÉ„ÅÆ„Çµ„Ç§„Ç∫„Åã„Çâ„Åù„Çå„Åû„Çå 0.3ÂÄç „Åó„Å¶„Åè„Å†„Åï„ÅÑ

</aside>

<aside>
‚ö†Ô∏è ‰øùÂ≠ò„Åô„ÇãÂãïÁîª„ÅÆ„Éï„Ç°„Ç§„É´Âêç„ÅØ Â≠¶Á±çÁï™Âè∑_cat_binary.mp4 „Å®„Åó„Å¶„Åè„Å†„Åï„ÅÑ

</aside>

```python
### ÔºîÔºé‰∫åÂÄ§Âåñ„Å®ÂãïÁîª„ÅÆ‰øùÂ≠òÔºàÊúÄÂ§ß10ÁÇπÔºâ

import cv2
import numpy as np

# „É≠„Éº„Ç´„É´Áí∞Â¢ÉÁî®
path = "Cat - 66004.mp4"
save_path = "6322087_cat_binary.mp4"  # ‰øùÂ≠ò„Éï„Ç°„Ç§„É´Âêç

# ÊèêÂá∫(google colabÁí∞Â¢É)
# path = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/Cat - 66004.mp4"
# save_path = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/output/6322087_cat_binary.mp4"

# ÂãïÁîª„Éï„Ç°„Ç§„É´„ÇíË™≠„ÅøËæº„Åø
video = cv2.VideoCapture(path)

# ÂãïÁîª„ÅÆ„Çµ„Ç§„Ç∫„ÇíÂèñÂæó
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))

# ‰øùÂ≠ò„Åô„ÇãÂãïÁîª„ÅÆ„Çµ„Ç§„Ç∫
width = int(width * 0.3)
height = int(height * 0.3)
print("ÂãïÁîª„ÅÆÂπÖ:", width)
print("ÂãïÁîª„ÅÆÈ´ò„Åï:", height)

# „Éï„Ç©„Éº„Éû„ÉÉ„Éà„ÄÅ„Ç≥„Éº„Éá„ÉÉ„ÇØ„ÄÅFPS„ÇíË®≠ÂÆö
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
fps = video.get(cv2.CAP_PROP_FPS)
writer = cv2.VideoWriter(save_path, fourcc, fps, (width, height), isColor=False)

# ÂãïÁîª„ÅÆ„Éï„É¨„Éº„É†„Çí1„Å§„Åö„Å§ÂèñÂæó„Åó„Å¶‰∫åÂÄ§Âåñ„Åó„ÄÅ‰øùÂ≠ò„Åô„Çã
while True:
    ret, frame = video.read()
    if not ret:
        break
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    _, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    binary = cv2.resize(binary, (width, height))
    writer.write(binary)

# „É°„É¢„É™„ÇíËß£Êîæ
video.release()
writer.release()

print("ÂãïÁîª„ÅÆÂá¶ÁêÜ„ÅåÂÆå‰∫Ü„Åó„Åæ„Åó„Åü„ÄÇ")
```

### Ôºï.„ÄÄÈü≥Â£∞„ÅÆÂèØË¶ñÂåñÔºàÊúÄÂ§ß10ÁÇπÔºâ

Èü≥Â£∞Ôºàreport_audio.wavÔºâ„Å´„ÅØÊôÇÈñì-Âë®Ê≥¢Êï∞È†òÂüü„Å´„É°„ÉÉ„Çª„Éº„Ç∏„ÅåÈö†„Åï„Çå„Å¶„ÅÑ„Åæ„ÅôÔºé

„Éë„É©„É°„Éº„Çø„ÇíÈÅ©ÂÆúË™øÁØÄ„Åó„Å¶Ëß£Êûê„ÅóÔºåÂèØË¶ñÂåñ„Åó„Åü„Ç∞„É©„Éï„ÇíÊèèÁîª„Åó„Å¶„Åè„Å†„Åï„ÅÑÔºé

```python
### Ôºï.„ÄÄÈü≥Â£∞„ÅÆÂèØË¶ñÂåñÔºàÊúÄÂ§ß10ÁÇπÔºâ
# Èü≥Â£∞Ôºàreport_audio.wavÔºâ„Å´„ÅØÊôÇÈñì-Âë®Ê≥¢Êï∞È†òÂüü„Å´„É°„ÉÉ„Çª„Éº„Ç∏„ÅåÈö†„Åï„Çå„Å¶„ÅÑ„Åæ„ÅôÔºé
# „Éë„É©„É°„Éº„Çø„ÇíÈÅ©ÂÆúË™øÁØÄ„Åó„Å¶Ëß£Êûê„ÅóÔºåÂèØË¶ñÂåñ„Åó„Åü„Ç∞„É©„Éï„ÇíÊèèÁîª„Åó„Å¶„Åè„Å†„Åï„ÅÑÔºé

import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display

# „É≠„Éº„Ç´„É´Áí∞Â¢ÉÁî®
path = "report_audio.wav"

# ÊèêÂá∫(google colabÁí∞Â¢É)Áî®
# path = "/content/drive/MyDrive/Colab Notebooks/dataset/audios/report_audio.wav"

# Èü≥Â£∞„Éï„Ç°„Ç§„É´„ÇíË™≠„ÅøËæº„Åø
y, sr = librosa.load(path)

print(sr)

nfft = 1024
D = librosa.amplitude_to_db(np.abs(librosa.stft(y, hop_length=1024, n_fft = nfft)))
fig, ax = plt.subplots(figsize=(12, 3))
img = librosa.display.specshow(D, y_axis='linear', sr=sr, hop_length=1024, x_axis='time', cmap='viridis', ax=ax)
plt.colorbar(img, ax=ax)

# df = sr / N
print("D.shape:", D.shape)
print("df:", sr/nfft)
```

```python
### ÔºñÔºéÂãïÁîª„ÅÆÂ§â„Çè„ÇäÁõÆ„ÇíÊ§úÂá∫„Åô„ÇãÔºàÊúÄÂ§ß25ÁÇπÔºâ

import cv2
import numpy as np
import matplotlib.pyplot as plt

# „É≠„Éº„Ç´„É´Áí∞Â¢ÉÁî®
video_path = "Detection.mp4"

# ÊèêÂá∫(google colabÁí∞Â¢É)Áî®
# video_path = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/Detection.mp4"

# ÂãïÁîª„ÇíË™≠„ÅøËæº„Åø ‚ÜìÂãïÁîª„ÅÆË™≠„ÅøËæº„Åø„ÅØopenCV„Çí‰Ωø„ÅÜ„Åó„Åã„Å™„ÅÑ„Åü„ÇÅOKÔºü
video = cv2.VideoCapture(video_path)

# „Éí„Çπ„Éà„Ç∞„É©„É†„ÅÆbinÊï∞
hist_bins = 64

# Áõ∏Èñ¢„ÅÆ„É™„Çπ„Éà„ÇíÂàùÊúüÂåñ
correlations = []

#„Ç∞„É¨„Ç§„Çπ„Ç±„Éº„É´„Å´Â§âÊèõ„Åô„ÇãÈñ¢Êï∞
def rgb2gray(frame):
    gray = frame[:,:,0] * 0.299 + frame[:,:,1] * 0.587 + frame[:,:,2] * 0.114
    gray = gray.astype(np.uint8)
    return gray

# „Éí„Çπ„Éà„Ç∞„É©„É†„ÅÆË®àÁÆó„Çí„Åô„ÇãÈñ¢Êï∞
def calc_hist(frame, hist_bins, hist_range):
    # „Éí„Çπ„Éà„Ç∞„É©„É†„ÅÆÂàùÊúüÂåñ
    hist = np.zeros(hist_bins, dtype=int)
    bin_width = (hist_range[1] - hist_range[0]) / hist_bins
    # ÁîªÂÉè„ÅÆÂêÑÁîªÁ¥†ÂÄ§„Çí„Éí„Çπ„Éà„Ç∞„É©„É†„Å´ËøΩÂä†
    for i in range(hist_bins):
        # ÊúÄÂæå„ÅÆ„Éì„É≥„ÅÆ‰∏äÈôê„ÅØhist_range[1]„Å®„Åô„Çã
        bin_upper = hist_range[1] if i == hist_bins - 1 else hist_range[0] + (i + 1) * bin_width
        hist[i] = np.sum((frame >= hist_range[0] + i * bin_width) & (frame < bin_upper))
    return hist

# „Éí„Çπ„Éà„Ç∞„É©„É†Èñì„ÅÆÁõ∏Èñ¢„ÇíË®àÁÆó„Åô„ÇãÈñ¢Êï∞
def compare_hist(hist1, hist2):
    # „Éí„Çπ„Éà„Ç∞„É©„É†„ÅÆÂπ≥ÂùáÂÄ§„ÇíË®àÁÆó
    mean1 = np.mean(hist1)
    mean2 = np.mean(hist2)
    # „Éí„Çπ„Éà„Ç∞„É©„É†„ÅÆÂàÜÊï£„ÇíË®àÁÆó
    var1 = np.var(hist1)
    var2 = np.var(hist2)
    # „Éí„Çπ„Éà„Ç∞„É©„É†„ÅÆÂÖ±ÂàÜÊï£„ÇíË®àÁÆó
    cov = np.mean((hist1 - mean1) * (hist2 - mean2))
    # Áõ∏Èñ¢‰øÇÊï∞„ÇíË®àÁÆó
    correlation = cov / np.sqrt(var1 * var2)
    return correlation

# ÊúÄÂàù„ÅÆ„Éï„É¨„Éº„É†„ÇíË™≠„ÅøËæº„Åø
ret, prev_frame = video.read()
prev_gray = rgb2gray(prev_frame)
prev_hist = calc_hist(prev_gray, hist_bins, [0, 255])

# „Éï„É¨„Éº„É†„Åî„Å®„Å´Âá¶ÁêÜ
frame_count = 1
while True:
    ret, frame = video.read()
    if not ret:
        break

    # „Ç∞„É¨„Éº„Çπ„Ç±„Éº„É´„Å´Â§âÊèõ
    gray = rgb2gray(frame)

    # „Éí„Çπ„Éà„Ç∞„É©„É†„ÇíË®àÁÆó
    hist = calc_hist(gray, hist_bins, [0, 255])

    # „Éí„Çπ„Éà„Ç∞„É©„É†„ÅÆÁõ∏Èñ¢„ÇíË®àÁÆó
    correlation = compare_hist(prev_hist, hist)
    correlations.append(correlation)

    # ÁèæÂú®„ÅÆ„Éï„É¨„Éº„É†„ÇíÊ¨°„ÅÆ„Éï„É¨„Éº„É†„ÅÆÊØîËºÉÁî®„Å´‰øùÂ≠ò
    prev_hist = hist

    frame_count += 1

# ÂãïÁîª„ÇíËß£Êîæ
video.release()

# Áõ∏Èñ¢„Çí„Ç∞„É©„Éï„Å´„Éó„É≠„ÉÉ„Éà
plt.figure(figsize=(12, 6))
plt.plot(correlations)
plt.xlabel("flame number") # „Éï„É¨„Éº„É†Áï™Âè∑
plt.ylabel("correlation") # Áõ∏Èñ¢
plt.title("Correlation between consecutive frames") # ÈÄ£Á∂ö„Åô„Çã„Éï„É¨„Éº„É†Èñì„ÅÆÁõ∏Èñ¢
plt.show()

# ÈñæÂÄ§„ÇíË®≠ÂÆö„Åó„Å¶„Ç∑„Éº„É≥„ÉÅ„Çß„É≥„Ç∏„ÇíÊ§úÂá∫
threshold = 0.9  # ÈñæÂÄ§„ÇíË®≠ÂÆö
scene_changes = np.where(np.array(correlations) < threshold)[0]
print("„Ç∑„Éº„É≥„ÉÅ„Çß„É≥„Ç∏„ÅÆ„Éï„É¨„Éº„É†Áï™Âè∑:", scene_changes)
```

```python
### ÔºóÔºé„Ç™„Éñ„Ç∏„Çß„ÇØ„ÉàÊåáÂêëÔºöVideo „ÇØ„É©„Çπ„ÅÆ‰ΩúÊàêÔºàÊúÄÂ§ß25ÁÇπÔºâ

import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

class Video:
    def __init__(self, path):
        self.path = path
        self.create_title_and_id()
        self.create_thumbnail()

    # „Çµ„É†„Éç„Ç§„É´„Çí‰ΩúÊàê„Åô„Çã„É°„ÇΩ„ÉÉ„Éâ„Çí‰ΩúÊàê
    def create_thumbnail(self):
        video = cv2.VideoCapture(self.path)
        ret, frame = video.read()
        video.release()
        if ret:
            self.thumbnail = frame
        else:
            print(f"„Çµ„É†„Éç„Ç§„É´„ÅÆ‰ΩúÊàê„Å´Â§±Êïó„Åó„Åæ„Åó„Åü„ÄÇ")
            self.thumbnail = None

    # „Éï„Ç°„Ç§„É´Âêç„Åã„ÇâÁâπÂÆö„ÅÆÊñáÂ≠óÂàóÔºà„Çø„Ç§„Éà„É´„Å®idÔºâ„ÇíÂèñ„ÇäÂá∫„Åô„É°„ÇΩ„ÉÉ„Éâ„ÅÆ‰ΩúÊàê
    def create_title_and_id(self):
        try:
            file_name = os.path.basename(self.path)
            title, id = file_name.split(" - ")
            id = id.split(".")[0]
            self.title = title
            self.id = id
        except ValueError:
            print(f"„Éï„Ç°„Ç§„É´Âêç({file_name})„Åå‰∏çÊ≠£„Åß„Åô„ÄÇ")
            return None, None
    
    # Ôºî„Å§„ÅÆ„Ç§„É≥„Çπ„Çø„É≥„ÇπÂ§âÊï∞„ÅÆÂÄ§„ÇíÂèñÂæó„Åô„ÇãÔºî„Å§„ÅÆ„É°„ÇΩ„ÉÉ„ÉâÔºà„Ç≤„ÉÉ„Çø„ÉºÔºâ
    def get_id(self):
        return self.id
    
    def get_path(self):
        return self.path
    
    def get_title(self):
        return self.title
    
    def get_thumbnail(self):
        return self.thumbnail
    
    # ÔºíÂÄ§Âåñ„Åô„ÇãÂãïÁîª„Çí‰ΩúÊàê„Åô„Çã„É°„ÇΩ„ÉÉ„Éâ„ÅÆ‰ΩúÊàê
    def to_binary(self, save_path):
        video = cv2.VideoCapture(self.path)
        width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
        fps = video.get(cv2.CAP_PROP_FPS)
        width = int(width * 0.3)
        height = int(height * 0.3)
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        writer = cv2.VideoWriter(save_path, fourcc, fps, (width, height), isColor=False)
        while True:
            ret, frame = video.read()
            if not ret:
                break
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            _, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
            binary = cv2.resize(binary, (width, height))
            writer.write(binary)
        video.release()
        writer.release()
        print("ÂãïÁîª„ÅÆÂá¶ÁêÜ„ÅåÂÆå‰∫Ü„Åó„Åæ„Åó„Åü„ÄÇ")

    # „Éï„É¨„Éº„É†„ÇíÔºíÂÄ§Âåñ„Åô„Çã„É°„ÇΩ„ÉÉ„Éâ
    def binarize_frame(self, frame):
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        _, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
        return binary
    
    # „Éï„É¨„Éº„É†„Çí„Ç∞„É¨„Éº„Çπ„Ç±„Éº„É´Âåñ„Åô„Çã„É°„ÇΩ„ÉÉ„Éâ
    def rgb2gray(self, frame):
        gray = frame[:,:,0] * 0.299 + frame[:,:,1] * 0.587 + frame[:,:,2] * 0.114
        gray = gray.astype(np.uint8)
        return gray

    # „Éï„É¨„Éº„É†Âá¶ÁêÜ„É°„ÇΩ„ÉÉ„Éâ„ÇíÂèó„ÅëÂèñ„Çä„ÄÅ„Éï„É¨„Éº„É†Âá¶ÁêÜ„ÇíÂãïÁîªÂÖ®‰Ωì„Å´ÊñΩ„Åô„É°„ÇΩ„ÉÉ„Éâ
    def process_video(self, func, save_path):
        video = cv2.VideoCapture(self.path)
        width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
        fps = video.get(cv2.CAP_PROP_FPS)
        width = int(width * 0.3)
        height = int(height * 0.3)
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        writer = cv2.VideoWriter(save_path, fourcc, fps, (width, height), isColor=False)
        while True:
            ret, frame = video.read()
            if not ret:
                break
            processed_frame = func(frame)
            processed_frame = cv2.resize(processed_frame, (width, height))
            writer.write(processed_frame)
        video.release()
        writer.release()
        print("ÂãïÁîª„ÅÆÂá¶ÁêÜ„ÅåÂÆå‰∫Ü„Åó„Åæ„Åó„Åü„ÄÇ")

# „ÉÜ„Çπ„Éà
# „É≠„Éº„Ç´„É´Áí∞Â¢ÉÁî®
path = "People - 84973.mp4"
save_path = "./"

# ÊèêÂá∫(google colabÁí∞Â¢É)Áî®
# path = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/People - 84973.mp4"
# save_path = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/test/"

video = Video(path)

# 1.„Çµ„É†„Éç„Ç§„É´„ÅåÊ≠£„Åó„Åè‰ΩúÊàê„Åï„Çå„Å¶„ÅÑ„Çã„ÅãÁ¢∫Ë™ç
print("1. „Çµ„É†„Éç„Ç§„É´„ÅåÊ≠£„Åó„Åè‰ΩúÊàê„Åï„Çå„Å¶„ÅÑ„Çã„ÅãÁ¢∫Ë™ç")
plt.imshow(video.thumbnail)
plt.show()

# 2.„Éï„Ç°„Ç§„É´Âêç„Åã„Çâ„Çø„Ç§„Éà„É´„Å®ID„ÅåÊ≠£„Åó„ÅèÂèñÂæó„Åß„Åç„Å¶„ÅÑ„Çã„ÅãÁ¢∫Ë™ç
print("2. „Éï„Ç°„Ç§„É´Âêç„Åã„Çâ„Çø„Ç§„Éà„É´„Å®ID„ÅåÊ≠£„Åó„ÅèÂèñÂæó„Åß„Åç„Å¶„ÅÑ„Çã„ÅãÁ¢∫Ë™ç")
print("Title:", video.title)
print("ID:", video.id)

# 3~5. 4„Å§„ÅÆ„Ç§„É≥„Çπ„Çø„É≥„ÇπÂ§âÊï∞„ÅÆÂÄ§„ÇíÂèñÂæó„Åô„Çã4„Å§„ÅÆ„É°„ÇΩ„ÉÉ„ÉâÔºà„Ç≤„ÉÉ„Çø„ÉºÔºâ„ÅÆÁ¢∫Ë™ç
print("3~5. 4„Å§„ÅÆ„Ç§„É≥„Çπ„Çø„É≥„ÇπÂ§âÊï∞„ÅÆÂÄ§„ÇíÂèñÂæó„Åô„Çã4„Å§„ÅÆ„É°„ÇΩ„ÉÉ„ÉâÔºà„Ç≤„ÉÉ„Çø„ÉºÔºâ„ÅÆÁ¢∫Ë™ç")
print("ID:", video.get_id())
print("Path:", video.get_path())
print("Title:", video.get_title())
print("Thumbnail:")
plt.imshow(video.get_thumbnail())
plt.show()

# 6. ÔºíÂÄ§Âåñ„Åô„ÇãÂãïÁîª„Çí‰ΩúÊàê„Åô„Çã„É°„ÇΩ„ÉÉ„Éâ„ÅÆÁ¢∫Ë™ç
print("6. ÔºíÂÄ§Âåñ„Åô„ÇãÂãïÁîª„Çí‰ΩúÊàê„Åô„Çã„É°„ÇΩ„ÉÉ„Éâ„ÅÆÁ¢∫Ë™ç")
video.to_binary(save_path + "6322087_People_binary.mp4")
# „Çµ„É†„Éç„Çí‰Ωú„Å£„Å¶Á¢∫Ë™ç
binarized_video = Video(save_path + "6322087_People_binary.mp4")
plt.imshow(binarized_video.get_thumbnail())
plt.show()

# 7. „Éï„É¨„Éº„É†Âá¶ÁêÜ„É°„ÇΩ„ÉÉ„Éâ„ÇíÂèó„ÅëÂèñ„Çä„ÄÅ„Éï„É¨„Éº„É†Âá¶ÁêÜ„ÇíÂãïÁîªÂÖ®‰Ωì„Å´ÊñΩ„Åô„É°„ÇΩ„ÉÉ„Éâ„ÅÆÁ¢∫Ë™ç
print("7. „Éï„É¨„Éº„É†Âá¶ÁêÜ„É°„ÇΩ„ÉÉ„Éâ„ÇíÂèó„ÅëÂèñ„Çä„ÄÅ„Éï„É¨„Éº„É†Âá¶ÁêÜ„ÇíÂãïÁîªÂÖ®‰Ωì„Å´ÊñΩ„Åô„É°„ÇΩ„ÉÉ„Éâ„ÅÆÁ¢∫Ë™ç")
print("7.1. ‰∫åÂÄ§Âåñ„ÅÆÂãïÁîªÂá¶ÁêÜ„Å´„Å§„ÅÑ„Å¶Á¢∫Ë™ç")
video.process_video(video.binarize_frame, save_path + "6322087_People_binary2.mp4")
binarized_video2 = Video(save_path + "6322087_People_binary2.mp4")
plt.imshow(binarized_video2.get_thumbnail())
plt.show()

print("7.2. „Ç∞„É¨„Éº„Çπ„Ç±„Éº„É´Âåñ„ÅÆÂãïÁîªÂá¶ÁêÜ„Å´„Å§„ÅÑ„Å¶Á¢∫Ë™ç")
video.process_video(video.rgb2gray, save_path + "6322087_People_gray.mp4")
gray_video = Video(save_path + "6322087_People_gray.mp4")
plt.imshow(gray_video.get_thumbnail(), cmap="gray")
plt.show()
```
