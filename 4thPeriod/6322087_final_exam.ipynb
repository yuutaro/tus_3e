### ４．二値化と動画の保存（最大10点）

動画（Cat - 66004.mp4）を２値化した動画を作成するプログラムを作成し，その動画を保存してください．閾値は自分で設定してかまいません．

二値化：画素値が閾値より大きければ白を割り当て，そうでなければ黒を割り当てる処理

<aside>
⚠️ 縦・横のサイズは元のサイズからそれぞれ 0.3倍 してください

</aside>

<aside>
⚠️ 保存する動画のファイル名は 学籍番号_cat_binary.mp4 としてください

</aside>

```python
### ４．二値化と動画の保存（最大10点）

import cv2
import numpy as np

# ローカル環境用
path = "Cat - 66004.mp4"
save_path = "6322087_cat_binary.mp4"  # 保存ファイル名

# 提出(google colab環境)
# path = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/Cat - 66004.mp4"
# save_path = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/output/6322087_cat_binary.mp4"

# 動画ファイルを読み込み
video = cv2.VideoCapture(path)

# 動画のサイズを取得
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))

# 保存する動画のサイズ
width = int(width * 0.3)
height = int(height * 0.3)
print("動画の幅:", width)
print("動画の高さ:", height)

# フォーマット、コーデック、FPSを設定
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
fps = video.get(cv2.CAP_PROP_FPS)
writer = cv2.VideoWriter(save_path, fourcc, fps, (width, height), isColor=False)

# 動画のフレームを1つずつ取得して二値化し、保存する
while True:
    ret, frame = video.read()
    if not ret:
        break
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    _, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    binary = cv2.resize(binary, (width, height))
    writer.write(binary)

# メモリを解放
video.release()
writer.release()

print("動画の処理が完了しました。")
```

### ６．動画の変わり目を検出する（最大25点）

動画（Detection.mp4）は４つの動画を結合したものです．

この動画から動画の変わり目（３つ）を以下の方法で自動で検出してください．

動画の変わり目は，前のフレームと比較をして**画素値の分布**の変化量が多いフレームといえます．

つまり，

1. 各フレームのヒストグラムを計算する
2. t 番目のフレームと t-1 番目のフレーム間のヒストグラムの相関を計算する

これらを全てのフレームに対して行い，相関が弱いフレームを見つけることで実現できます．

a. 横軸が時間（フレーム），縦軸が相関のグラフを描画してください（12点）．

b. 変化量がある値（閾値）よりも小さいフレームを動画の変わり目としたとき，何番目のフレームが動画の変わり目かを出力してください．閾値は自分で設定してかまいません．（12点）

定義はopenCVのcompareHist()を参考にすると良いと思います．

cf. [compareHist()](http://opencv.jp/opencv-2svn/cpp/imgproc_histograms.html)

考察の余地：

- ヒストグラムのbin
- 閾値
- 画素間の分布の変化量の尺度
    - KLダイバージェンス etc.

<aside>
⚠️ この課題ではcompareHist()やcalcHist()といったopenCVの関数は使用してはいけません．
numpy等で実装することを期待しています

</aside>

```python
### ６．動画の変わり目を検出する（最大25点）

import cv2
import numpy as np
import matplotlib.pyplot as plt

# ローカル環境用
video_path = "Detection.mp4"

# 提出(google colab環境)用
# video_path = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/Detection.mp4"

# 動画を読み込み ↓動画の読み込みはopenCVを使うしかないためOK？
video = cv2.VideoCapture(video_path)

# ヒストグラムのbin数
hist_bins = 64

# 相関のリストを初期化
correlations = []

#グレイスケールに変換する関数
def rgb2gray(frame):
    gray = frame[:,:,0] * 0.299 + frame[:,:,1] * 0.587 + frame[:,:,2] * 0.114
    gray = gray.astype(np.uint8)
    return gray

# ヒストグラムの計算をする関数
def calc_hist(frame, hist_bins, hist_range):
    # ヒストグラムの初期化
    hist = np.zeros(hist_bins, dtype=int)
    bin_width = (hist_range[1] - hist_range[0]) / hist_bins
    # 画像の各画素値をヒストグラムに追加
    for i in range(hist_bins):
        # 最後のビンの上限はhist_range[1]とする
        bin_upper = hist_range[1] if i == hist_bins - 1 else hist_range[0] + (i + 1) * bin_width
        hist[i] = np.sum((frame >= hist_range[0] + i * bin_width) & (frame < bin_upper))
    return hist

# ヒストグラム間の相関を計算する関数
def compare_hist(hist1, hist2):
    # ヒストグラムの平均値を計算
    mean1 = np.mean(hist1)
    mean2 = np.mean(hist2)
    # ヒストグラムの分散を計算
    var1 = np.var(hist1)
    var2 = np.var(hist2)
    # ヒストグラムの共分散を計算
    cov = np.mean((hist1 - mean1) * (hist2 - mean2))
    # 相関係数を計算
    correlation = cov / np.sqrt(var1 * var2)
    return correlation

# 最初のフレームを読み込み
ret, prev_frame = video.read()
prev_gray = rgb2gray(prev_frame)
prev_hist = calc_hist(prev_gray, hist_bins, [0, 255])

# フレームごとに処理
frame_count = 1
while True:
    ret, frame = video.read()
    if not ret:
        break

    # グレースケールに変換
    gray = rgb2gray(frame)

    # ヒストグラムを計算
    hist = calc_hist(gray, hist_bins, [0, 255])

    # ヒストグラムの相関を計算
    correlation = compare_hist(prev_hist, hist)
    correlations.append(correlation)

    # 現在のフレームを次のフレームの比較用に保存
    prev_hist = hist

    frame_count += 1

# 動画を解放
video.release()

# 相関をグラフにプロット
plt.figure(figsize=(12, 6))
plt.plot(correlations)
plt.xlabel("flame number") # フレーム番号
plt.ylabel("correlation") # 相関
plt.title("Correlation between consecutive frames") # 連続するフレーム間の相関
plt.show()

# 閾値を設定してシーンチェンジを検出
threshold = 0.9  # 閾値を設定
scene_changes = np.where(np.array(correlations) < threshold)[0]
print("シーンチェンジのフレーム番号:", scene_changes)
```

### ７．オブジェクト指向：Video クラスの作成（最大25点）

Videoクラスは 動画のpathを受け取り，インスタンス化する．

Videoクラスだけではなく，インスタンス化して動作を確認するプログラムもあわせて作成すること．

インスタンス変数として以下の４つを保持する：

- 動画の path
- タイトル
- ID
- サムネイル

コンストラクタ：

- 各変数を初期化

~~クラスメソッド~~ インスタンスメソッドとして以下の８つを用意する：

- サムネイルを作成するメソッドを作成
    - 課題２をクラス内のメソッドにする
    - サムネイルはインスタンス変数に代入
- ファイル名から特定の文字列（タイトルとID）を取り出すメソッドの作成
    - 課題３をクラス内のメソッドにする
    - タイトルとIDはインスタンス変数に代入
- ４つのインスタンス変数の値を取得する４つのメソッド（ゲッター）
    - get_title()
    - get_id()
    - get_thumbnail()
    - get_path()
- ２値化する動画を作成するメソッドの作成
    - 課題４をクラス内のメソッドにする
- ２値化する動画を作成するメソッドをフレームを処理する関数と動画処理を行う関数に分離し，引数としてフレームを処理する関数を受け取るように変更してください
    - cf. [高階関数](https://www.notion.so/1-Python-fa98e2f92db543c1b0a2f325d2abd760?pvs=21)
    
    （画像処理をする関数を増やしたとしても，その関数を渡すだけで動画処理ができるようにする）
    
    - 引数として受け取ったフレームに対して二値化する関数を作成
    - 引数として受け取ったフレームに対してグレースケール化する関数を作成
    - 上の２つの関数のどちらかを引数として受け取り，その関数が適用された動画を生成する関数を作成する

```python
### ７．オブジェクト指向：Video クラスの作成（最大25点）

import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

class Video:
    def __init__(self, path):
        self.path = path
        self.create_title_and_id()
        self.create_thumbnail()

    # サムネイルを作成するメソッドを作成
    def create_thumbnail(self):
        video = cv2.VideoCapture(self.path)
        ret, frame = video.read()
        video.release()
        if ret:
            self.thumbnail = frame
        else:
            print(f"サムネイルの作成に失敗しました。")
            self.thumbnail = None

    # ファイル名から特定の文字列（タイトルとid）を取り出すメソッドの作成
    def create_title_and_id(self):
        try:
            file_name = os.path.basename(self.path)
            title, id = file_name.split(" - ")
            id = id.split(".")[0]
            self.title = title
            self.id = id
        except ValueError:
            print(f"ファイル名({file_name})が不正です。")
            return None, None
    
    # ４つのインスタンス変数の値を取得する４つのメソッド（ゲッター）
    def get_id(self):
        return self.id
    
    def get_path(self):
        return self.path
    
    def get_title(self):
        return self.title
    
    def get_thumbnail(self):
        return self.thumbnail
    
    # ２値化する動画を作成するメソッドの作成
    def to_binary(self, save_path):
        video = cv2.VideoCapture(self.path)
        width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
        fps = video.get(cv2.CAP_PROP_FPS)
        width = int(width * 0.3)
        height = int(height * 0.3)
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        writer = cv2.VideoWriter(save_path, fourcc, fps, (width, height), isColor=False)
        while True:
            ret, frame = video.read()
            if not ret:
                break
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            _, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
            binary = cv2.resize(binary, (width, height))
            writer.write(binary)
        video.release()
        writer.release()
        print("動画の処理が完了しました。")

    # フレームを２値化するメソッド
    def binarize_frame(self, frame):
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        _, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
        return binary
    
    # フレームをグレースケール化するメソッド
    def rgb2gray(self, frame):
        gray = frame[:,:,0] * 0.299 + frame[:,:,1] * 0.587 + frame[:,:,2] * 0.114
        gray = gray.astype(np.uint8)
        return gray

    # フレーム処理メソッドを受け取り、フレーム処理を動画全体に施すメソッド
    def process_video(self, func, save_path):
        video = cv2.VideoCapture(self.path)
        width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
        fps = video.get(cv2.CAP_PROP_FPS)
        width = int(width * 0.3)
        height = int(height * 0.3)
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        writer = cv2.VideoWriter(save_path, fourcc, fps, (width, height), isColor=False)
        while True:
            ret, frame = video.read()
            if not ret:
                break
            processed_frame = func(frame)
            processed_frame = cv2.resize(processed_frame, (width, height))
            writer.write(processed_frame)
        video.release()
        writer.release()
        print("動画の処理が完了しました。")

# テスト
# ローカル環境用
path = "People - 84973.mp4"
save_path = "./"

# 提出(google colab環境)用
# path = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/People - 84973.mp4"
# save_path = "/content/drive/MyDrive/Colab Notebooks/dataset/videos/test/"

video = Video(path)

# 1.サムネイルが正しく作成されているか確認
print("1. サムネイルが正しく作成されているか確認")
plt.imshow(video.thumbnail)
plt.show()

# 2.ファイル名からタイトルとIDが正しく取得できているか確認
print("2. ファイル名からタイトルとIDが正しく取得できているか確認")
print("Title:", video.title)
print("ID:", video.id)

# 3~5. 4つのインスタンス変数の値を取得する4つのメソッド（ゲッター）の確認
print("3~5. 4つのインスタンス変数の値を取得する4つのメソッド（ゲッター）の確認")
print("ID:", video.get_id())
print("Path:", video.get_path())
print("Title:", video.get_title())
print("Thumbnail:")
plt.imshow(video.get_thumbnail())
plt.show()

# 6. ２値化する動画を作成するメソッドの確認
print("6. ２値化する動画を作成するメソッドの確認")
video.to_binary(save_path + "6322087_People_binary.mp4")
# サムネを作って確認
binarized_video = Video(save_path + "6322087_People_binary.mp4")
plt.imshow(binarized_video.get_thumbnail())
plt.show()

# 7. フレーム処理メソッドを受け取り、フレーム処理を動画全体に施すメソッドの確認
print("7. フレーム処理メソッドを受け取り、フレーム処理を動画全体に施すメソッドの確認")
print("7.1. 二値化の動画処理について確認")
video.process_video(video.binarize_frame, save_path + "6322087_People_binary2.mp4")
binarized_video2 = Video(save_path + "6322087_People_binary2.mp4")
plt.imshow(binarized_video2.get_thumbnail())
plt.show()

print("7.2. グレースケール化の動画処理について確認")
video.process_video(video.rgb2gray, save_path + "6322087_People_gray.mp4")
gray_video = Video(save_path + "6322087_People_gray.mp4")
plt.imshow(gray_video.get_thumbnail(), cmap="gray")
plt.show()
```
